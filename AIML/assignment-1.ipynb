# Task 2: Data Cleaning & Feature Engineering

### Checking for Missing Values

```python
df.isnull().sum()
```

### Handling the Missing Values

```python
df['country'] = df['country'].fillna('Unknown')
df['director'] = df['director'].fillna('No Director Listed')
df.isnull().sum()
```

### Duration Columns

```python
df['duration_minutes'] = float('nan')
df['seasons'] = float('nan')
df.loc[df['type'] == 'Movie', 'duration_minutes'] = df.loc[df['type'] == 'Movie', 'duration'].str.replace(' min', '').astype(float)
df.loc[df['type'] == 'TV Show', 'seasons'] = df.loc[df['type'] == 'TV Show', 'duration'].str.replace(' Season', '').str.replace('s', '').astype(float)
```

```python
df[['type', 'duration', 'duration_minutes', 'seasons']].head(10)
```

### Recent Release Flag

```python
df['Is_Recent'] = df['release_year'].apply(lambda x: 1 if x >= 2015 else 0)
df['Is_Recent'].value_counts()
```

```python
df.head()
```

# Task 3: Exploratory Data Analysis (EDA)

## Univariate Analysis
### Count plot of type (Movies vs TV Shows)
A count plot will help us compare the frequency of two different categories.

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
sns.countplot(x='type', data=df)
plt.title('Distribution of Movies vs TV Shows')
```

###  Histogram of release year
A Histogram will help us understand how the number of releases are distributed over ranges of years and reveal the trends over time.


```python
plt.subplot(1, 2, 2)
sns.histplot(df['release_year'], bins=30, kde=False)
plt.title('Distribution of Release Years')
plt.tight_layout()
plt.show()
```

## Bivariate analysis
### Bar plot: Number of releases per country
Bar plots are ideal for comparing counts across categories. We can identify which countries contribute the most in content production.

```python
plt.figure(figsize=(12, 6))
top_countries = df['country'].value_counts().head(10)
sns.barplot(x=top_countries.index, y=top_countries.values)
plt.title('Top 10 Countries by Content Production')
plt.xlabel('Country')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.show()
```

### Boxplot: Comparison between duration of movies (in minutes) from older years and recent years
Boxplots compare distributions between groups. Rather than relying only on averages for comparison, it helps to identify the spread and outliers.

```python
plt.figure(figsize=(8, 6))
movies_df = df[df['type'] == 'Movie']
sns.boxplot(x='Is_Recent', y='duration_minutes', data=movies_df)
plt.title('Comparison of Movie Duration: Recent (1) vs Older (0)')
plt.xlabel('Is Recent (1=Yes, 0=No)')
plt.ylabel('Duration (Minutes)')
plt.show()
```

## Text Analysis Insight

```python
from wordcloud import WordCloud
text = " ".join(title for title in df.listed_in.dropna())
word_cloud = WordCloud(collocations=False, background_color='white', width=800, height=400).generate(text)
```

```python
plt.figure(figsize=(10, 5))
plt.imshow(word_cloud, interpolation='bilinear')
plt.axis("off")
plt.title('Word Cloud of Genres')
plt.show()
```

## Correlation Heatmap
It helps in measuring linear relationships between numerical variables by allowing simultaneous comparisons in a compact form

```python
plt.figure(figsize=(8, 6))
numerical_cols = ['release_year', 'duration_minutes', 'seasons', 'Is_Recent']
sns.heatmap(df[numerical_cols].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```
